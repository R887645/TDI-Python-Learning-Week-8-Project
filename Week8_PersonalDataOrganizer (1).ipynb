```python
#Task 1 : Basic GroupBy by Single Column 

import pandas as pd

# Load your dataset
df = pd.read_csv("supply_chain_cleaned.csv")

# Group by supplier_name and calculate summary statistics
supplier_summary = (
    df.groupby("supplier_name")
      .agg(
          total_orders=("order_id", "count"),       # count of orders
          total_cost=("total_cost", "sum"),         # sum of total_cost
          average_cost=("total_cost", "mean"),      # mean of total_cost
          min_cost=("total_cost", "min"),           # minimum total_cost
          max_cost=("total_cost", "max")            # maximum total_cost
      )
      .sort_values(by="total_cost", ascending=False)  # sort by total cost
)

# Display results
print(supplier_summary)
```

```python
#Task 2: Group BY Mulitple Columns
import pandas as pd

# Load your dataset
df = pd.read_csv("supply_chain_cleaned.csv")

# Group by supplier_name and product_category
supplier_category_summary = (
    df.groupby(["supplier_name", "product_category"])
      .agg(
          order_count=("order_id", "count"),          # count of orders
          average_cost=("total_cost", "mean"),        # average cost per order
          total_quantity=("quantity", "sum")          # total quantity supplied
      )
      .sort_values(by=["supplier_name", "product_category"])  # sort by supplier & category
)

# Display results
print(supplier_category_summary)
```

```python
#Task 3: Calculate Multiple Metrics with Named Aggregations 

import pandas as pd

# Load your dataset
df = pd.read_csv("supply_chain_cleaned.csv")

# Group by supplier_name with multiple named aggregations
supplier_metrics = (
    df.groupby("supplier_name")
      .agg(
          total_orders=("order_id", "count"),       # count of orders
          avg_cost=("total_cost", "mean"),          # average cost
          std_cost=("total_cost", "std"),           # standard deviation of cost
          min_cost=("total_cost", "min"),           # minimum cost
          max_cost=("total_cost", "max")            # maximum cost
      )
      .sort_values(by="total_orders", ascending=False)  # sort by order volume
)

# Display top 3 suppliers by volume
print(supplier_metrics.head(3))
```

```python
#Task 4: Task 4: Create Pivot Tables 

import pandas as pd

# Load dataset
df = pd.read_csv("supply_chain_cleaned.csv")

# --- Pivot 1: Suppliers as rows, product_categories as columns, average cost as values
pivot_supplier_category = pd.pivot_table(
    df,
    index="supplier_name",
    columns="product_category",
    values="total_cost",
    aggfunc="mean"
)

print("Pivot: Supplier vs Product Category (Average Cost)")
print(pivot_supplier_category)

# --- Pivot 2: Product categories as rows, months as columns, total quantity as values
# First, ensure you have a 'month' column extracted from your date field
df["month"] = pd.to_datetime(df["order_date"]).dt.month

pivot_category_month = pd.pivot_table(
    df,
    index="product_category",
    columns="month",
    values="quantity",
    aggfunc="sum"
)

print("\nPivot: Product Category vs Month (Total Quantity)")
print(pivot_category_month)

# --- Cheapest supplier per product category
cheapest_suppliers = pivot_supplier_category.idxmin()
print("\nCheapest supplier for each product category:")
print(cheapest_suppliers)

# --- Highest volume product-month combinations
highest_volume = pivot_category_month.stack().idxmax()
print("\nProduct-Month combination with highest volume:")
print(highest_volume)
```

```python
#Task 5:Multi-Level Analysis 
import pandas as pd

# Load dataset
df = pd.read_csv("supply_chain_cleaned.csv")

# Ensure datetime conversion
df["order_date"] = pd.to_datetime(df["order_date"])
df["delivery_date"] = pd.to_datetime(df["delivery_date"])

# Calculate delivery_time in days
df["delivery_time"] = (df["delivery_date"] - df["order_date"]).dt.days

# ============================================================
# 1. On-time delivery rate by supplier
# ============================================================
if "delivery_status" in df.columns:
    on_time_rate = (
        df.groupby("supplier_name")["delivery_status"]
          .apply(lambda x: (x == "On Time").mean())
          .reset_index(name="on_time_rate")
    )
    print("1. On-time delivery rate by supplier:")
    print(on_time_rate)

# ============================================================
# 2. Average order value by product_category and delivery_status
# ============================================================
avg_order_value = (
    df.groupby(["product_category", "delivery_status"])["total_cost"]
      .mean()
      .reset_index(name="avg_order_value")
)
print("\n2. Average order value by product category & delivery status:")
print(avg_order_value)

# ============================================================
# 3. Delivery time trends by month
# ============================================================
df["month"] = df["order_date"].dt.month
delivery_trends = (
    df.groupby("month")["delivery_time"]
      .mean()
      .reset_index(name="avg_delivery_time")
)
print("\n3. Delivery time trends by month:")
print(delivery_trends)

# ============================================================
# 4. Identify underperforming supplier-product combinations
# ============================================================
if "delivery_status" in df.columns:
    supplier_product_perf = (
        df.groupby(["supplier_name", "product_category"])
          .agg(
              avg_cost=("total_cost", "mean"),
              on_time_rate=("delivery_status", lambda x: (x == "On Time").mean()),
              avg_delivery_time=("delivery_time", "mean")
          )
    )
    overall_mean_cost = df["total_cost"].mean()
    underperforming = supplier_product_perf[
        (supplier_product_perf["avg_cost"] > overall_mean_cost) &
        (supplier_product_perf["on_time_rate"] < 0.9)
    ]
    print("\n4. Underperforming supplier-product combinations:")
    print(underperforming)

# ============================================================
# 5. Supplier performance scorecard
# ============================================================
scorecard = (
    df.groupby("supplier_name")
      .agg(
          total_orders=("order_id", "count"),
          avg_cost=("total_cost", "mean"),
          on_time_rate=("delivery_status", lambda x: (x == "On Time").mean() if "delivery_status" in df.columns else None),
          avg_delivery_time=("delivery_time", "mean")
      )
)
print("\n5. Supplier Performance Scorecard:")
print(scorecard)

```
